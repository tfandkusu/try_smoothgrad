```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import tensorflow as tf
```

```python
model = tf.keras.models.load_model('model.hdf5')
labels = ['child', 'adult']
```

```python
child = cv2.imread('child.jpg')
child_xs = child.reshape((1, 200, 200, 3))
```

```python
def show(img):
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb)
```

```python
ys = model.predict(child_xs)
show(child)
print(labels[np.argmax(ys)])
```

```python
adult = cv2.imread('adult.jpg')
adult_xs = adult.reshape((1, 200, 200, 3))
```

```python
ys = model.predict(adult_xs)
show(adult)
print(labels[np.argmax(ys)])
```
